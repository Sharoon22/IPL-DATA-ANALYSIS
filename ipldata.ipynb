```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# id: Unique identifier for the match.
# season: The year and edition of the league (e.g., 2007/08).
# city: The city where the match was held.
# date: The date of the match.
# match_type: The type of match (e.g., League, Qualifier, Final).
# player_of_match: The player who was awarded the Player of the Match title.
# venue: The stadium or ground where the match took place.
# team1: The first team participating in the match.
# team2: The second team participating in the match.
# toss_winner: The team that won the toss.
# toss_decision: The decision made by the toss-winning team (bat or field).
# winner: The team that won the match.
# result: The outcome of the match (e.g., runs, wickets).
# result_margin: The margin by which the match was won (in runs or wickets).
# target_runs: The target score set for the chasing team.
# target_overs: The number of overs available to chase the target.
# super_over: Indicates whether a super over was played (Y/N).
# umpire1: The name of the first umpire.
# umpire2: The name of the second umpire.
```

## how many matches each team has won

```python
plt.figure(figsize=(10,5))
sns.countplot(x='winner',data=ipl)
plt.title('Match Results Distribution')
plt.xticks(rotation=90)
plt.ylabel('Number of Matches')
plt.show()
```

## which city most matches

```python
plt.figure(figsize=(12,6))
sns.countplot(x='city',palette='viridis',data=ipl)
plt.title('Matches Played by City')
plt.xticks(rotation=90)
plt.ylabel('Number of Matches')
plt.show()
```

## how many matches a team won in each season and performance over seasons
## compare teams who get stronger or weaker 

```python
plm=ipl.groupby(['season', 'winner']).size().reset_index(name='win')
```

```python
plm
```

```python
plt.figure(figsize=(12,5))
sns.barplot(data=plm, x='season',y='win', hue='winner')
```

## Player of the match

```python
pof=ipl['player_of_match'].value_counts().nlargest(10).reset_index()
```

```python
pof
```

```python
plt.figure(figsize=(11,6))
sns.barplot( x='player_of_match',y='count',data=pof)
plt.title('Player of match')
plt.ylabel('Number of Awards')
plt.xlabel('Player')
plt.show()
```

## Won based on runs

```python
runs=ipl.groupby('result')['winner'].get_group('runs').value_counts()
```

```python
run=pd.DataFrame(runs)
```

```python
run
```

```python
plt.figure(figsize=(11,6))
sns.lineplot(data=run,markers='d')
plt.title('won by runs')
plt.ylabel('Number of wins')
plt.xlabel('team')
plt.xticks(rotation=90)
plt.show()
```

## Won Based on wickets

```python
w=ipl.groupby('result')['winner'].get_group('wickets').value_counts()
```

```python
wicketss=pd.DataFrame(w)
```

```python
wicketss
```

```python
plt.figure(figsize=(11,6))
sns.lineplot(data=wicketss,markers='d')
plt.title('won by wickets')
plt.ylabel('Number of wins')
plt.xlabel('team')
plt.xticks(rotation=90)
plt.show()
```

## win based on their home ground

```python
hg=ipl.groupby('venue')['winner'].value_counts().nlargest(10)
```

```python
home=pd.DataFrame(hg)
```

```python
home
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='venue',y='count',hue='winner',palette='viridis',data=home)
plt.title('Matches Played in homeground and wins')
plt.xticks(rotation=90)
plt.ylabel('No of wins')
plt.show()
```
